```python
import matplotlib.pyplot as plt
import math
```

```python
xs = [0, 1, 2, 3]
ys = [0, 1, 2, 3]
ys2 = [3, 2, 1, 0]

plt.title('graph')
plt.plot(xs, ys, label='sample')
plt.plot(xs, ys2, label='sample')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
xs =[]
ys =[]
for i in range(1, 11):
    xs.append(i)
    ys.append(math.log(i))
    
plt.plot(xs, ys, label='$y = log(x)$')
plt.xlim([1, 10])
plt.title('Grapg of $y = log(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
